```python
%matplotlib inline

import numpy as np
from keras.datasets import mnist
from keras.preprocessing.image import ImageDataGenerator
from matplotlib import pyplot
```

```python
# load data
(X_train, y_train), (X_test, y_test) = mnist.load_data()
X_train = np.expand_dims(X_train, axis=3)
```

```python
def show_images_from_data_gen(datagen):
    datagen.fit(X_train)

    max_num_iterations = 1
    num_iterations = 0
    for X_batch, y_batch in datagen.flow(X_train, y_train, batch_size=9):

        if num_iterations >= max_num_iterations:
            break

        # Grid of 3x3 images
        for i in range(0, 9):
            pyplot.subplot(330 + 1 + i)
            pyplot.imshow(X_batch[i].reshape(28, 28), cmap=pyplot.get_cmap('gray'))
        pyplot.show()

        num_iterations += 1
```

```python
show_images_from_data_gen(ImageDataGenerator(rotation_range=45))
```

```python
# horizontal_flip: whether to randomly flip images horizontally.
# usually for real images, maybe not for digits!
show_images_from_data_gen(ImageDataGenerator(horizontal_flip=True))
```

```python
# zoom_range: amount of zoom. if scalar z, zoom will be randomly picked
#            in the range [1-z, 1+z]. A sequence of two can be passed instead
#            to select this range.
show_images_from_data_gen(ImageDataGenerator(zoom_range=0.5))
```

```python
# width_shift_range: fraction of total width.
# height_shift_range: fraction of total height.
show_images_from_data_gen(ImageDataGenerator(width_shift_range=0.4, height_shift_range=0.4))
```
